```python
import pandas as pd
```

```python
df_city = pd.read_csv('data/city_large.csv')
df_city
```

```python
df_city = df_city[df_city['City'].notnull()]
columns = {"latitude": [], 'longitude': []}
for i in df_city['Location'].iteritems():
    lat, lon =  i[1].split()
    
    columns['latitude'].append(float(lat[:-2]))
    columns['longitude'].append(-float(lon[:-2]))
df_temp = pd.DataFrame(columns)
```

```python
del df_city['Location']
```

```python
df_city = pd.concat([df_city, df_temp], axis=1)
```

```python
df_city = df_city.rename(columns={'2012 estimate': 'Population'})
```

```python
df_city.to_csv('data/city_large.csv', index=False)
```
